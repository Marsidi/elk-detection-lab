## Explore Mordor Dataset

### Initialize Analytics Engine

````
from openhunt.mordorutils import *
spark = get_spark()
````

### Download & Process Mordor File

````
mordor_file = "https://raw.githubusercontent.com/hunters-forge/mordor/master/datasets/small/windows/credential_access/covenant_mimikatz_logonpasswords.tar.gz"
registerMordorSQLTable(spark, mordor_file, "mordorTable")
````

### Get to know your data

````
df = spark.sql(
    '''
SELECT channel, COUNT(1)
FROM mordorTable
GROUP BY channel
    '''
)
df.show(10,False)
        
````
